```python
import pandas as pd
df = pd.read_csv('/content/cleaned_job_skills.csv', on_bad_lines='skip', engine='python')
print(df.isnull().sum())
print(df.duplicated().sum())
df = df.dropna(subset=['job_skills'])
df.drop_duplicates(inplace=True)
df['job_skills'].sample(10)
df.shape
df.head()
```

```python
#unique job titles and skill sets
print("Number of unique job titles:", df['job_link'].nunique())
print("Number of unique skill sets:", df['job_skills'].nunique())
```

```python
#min, max, avg skills per job
df['num_skills'] = df['job_skills'].apply(lambda x: len(x.split(',')))
df['num_skills'].describe()
```

```python
#most in demand skills
from collections import Counter
all_skills = df['job_skills'].dropna().str.split(', ')
flat_skills = [skill for sublist in all_skills for skill in sublist]
skill_counts = pd.Series(Counter(flat_skills)).sort_values(ascending=False)
print(skill_counts.head(20))
```
